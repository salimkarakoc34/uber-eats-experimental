```python
import mysql.connector
import json

db = mysql.connector.connect(
  host="xxxxxxxxxxxxxxxxx",
  user="xxxxxxxxxxxxxxxxx",
  password="xxxxxxxxxxxxxxxxx",  
  database="xxxxxxxxxxxxxxxxx"
)

cursor = db.cursor()
cursor.execute("CREATE TABLE hotels (street VARCHAR(255),  lat DECIMAL(10,8), lng DECIMAL(11,8),postalCode VARCHAR(255),state VARCHAR(255),price VARCHAR(255),address VARCHAR(255),title VARCHAR(255), ranking INT,neighborhood VARCHAR(255))")



with open('hotels.json',encoding='utf-8') as f:
    data = json.load(f)
    for item in data:
        street = item["street"]
        lat = item["location"]["lat"]
        lng = item["location"]["lng"]
        postalCode = item["postalCode"]
        state = item["state"]
        price = item["price"]
        address = item["address"]
        title = item["title"]
        ranking = item['rank']
        neighborhood = item["neighborhood"]



        sql = "INSERT INTO hotels (street, lat, lng, postalCode, state, price, address, title, ranking, neighborhood) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
        values = (street, lat, lng, postalCode, state, price, address, title, ranking, neighborhood)
        cursor.execute(sql, values)




db.commit()
db.close()
```

```python
import mysql.connector
import folium

db = mysql.connector.connect(
  host="xxxxxxxxxxxxxxxxx",
  user="xxxxxxxxxxxxxxxxx",
  password="xxxxxxxxxxxxxxxxx",  
  database="xxxxxxxxxxxxxxxxx"
)
cursor = db.cursor()

# Downtown Manhattan
query1 = "SELECT title, lat, lng FROM doordash WHERE  lat < 40.741670  "
query2 = "SELECT title, lat, lng FROM doordash WHERE NOT  lat < 40.741670  "


cursor.execute(query1)
data1 = cursor.fetchall()


cursor.execute(query2)
data2= cursor.fetchall()

manhattan_coords = [40.7831, -73.9712]

# Create the map
map = folium.Map(location=manhattan_coords, zoom_start=13)

for item in data1:
    name = item[0]
    coords = [item[1], item[2]]
    folium.Marker(location=coords, tooltip=name, icon=folium.Icon(color='cadetblue')).add_to(map)

for item in data2:
    name = item[0]
    coords = [item[1], item[2]]
    folium.Marker(location=coords, tooltip=name, icon=folium.Icon(color='green')).add_to(map)

# Show the map
map
```

```python
import random
db = mysql.connector.connect(
  host="xxxxxxxxxxxxxxxxx",
  user="xxxxxxxxxxxxxxxxx",
  password="xxxxxxxxxxxxxxxxx",  
  database="xxxxxxxxxxxxxxxxx"
)
cursor = db.cursor()
```

```python
# restaurants random selector 
cursor.execute("SELECT  lat, lng FROM doordash ORDER BY RAND() LIMIT 4")
res_locs = cursor.fetchall()
```

```python
cursor.execute("SELECT  lat, lng FROM doordash ORDER BY RAND() LIMIT 6")
delivery_guys = cursor.fetchall()
```

```python
# hotels random selector 
cursor.execute("SELECT lat, lng, title FROM hotels ORDER BY RAND() LIMIT 4")
hotel_locs = cursor.fetchall()
```

```python
manhattan_coords = [40.7831, -73.9712]

map = folium.Map(location=manhattan_coords, zoom_start=12)

# restaurant locations
for i in res_locs:
    folium.Marker(location=[i[0], i[1]], icon=folium.Icon(color='green')).add_to(map)

# delivery  guys  
for j in delivery_guys:
    folium.Marker(location=[j[0], j[1]], icon=folium.Icon(color='blue')).add_to(map)


# hotels locations
for k in hotel_locs:
    folium.Marker(location=[k[0], k[1]], tooltip=k[2], icon=folium.Icon(color='black',icon='bed')).add_to(map)


map
```
